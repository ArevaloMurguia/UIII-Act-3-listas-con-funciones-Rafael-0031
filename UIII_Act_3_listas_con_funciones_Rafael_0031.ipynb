## Soporte tecnico

```python
# Lista para guardar datos
nombres = []
equipos = []
problemas = []

# Función agregar
def agregarSoporte():
    nombre = input("Nombre del cliente: ")
    equipo = input("Equipo: ")
    problema = input("Problema: ")

    nombres.append(nombre)
    equipos.append(equipo)
    problemas.append(problema)

    print("Reporte guardado")


# Función leer
def leerSoporte():
    nombre = input("Escribe tu nombre: ")
    equipo = input("Escribe tu equipo: ")
    problema = input("Escribe tu problema: ")

    print("Cliente:", nombre)
    print("Equipo:", equipo)
    print("Problema:", problema)


# Función listar
def listarSoporte():
    if len(nombres) == 0:
        print("No hay reportes")
    else:
        i = 0
        while i < len(nombres):
            print("Reporte", i + 1)
            print("Cliente:", nombres[i])
            print("Equipo:", equipos[i])
            print("Problema:", problemas[i])
            i = i + 1


# Función buscar
def buscarSoporte():
    buscar = input("Nombre a buscar: ")

    i = 0
    encontrado = False

    while i < len(nombres):
        if nombres[i] == buscar:
            print("Cliente:", nombres[i])
            print("Equipo:", equipos[i])
            print("Problema:", problemas[i])
            encontrado = True
        i = i + 1

    if encontrado == False:
        print("No encontrado")


# Función modificar
def modificarSoporte():
    buscar = input("Nombre a modificar: ")

    i = 0

    while i < len(nombres):
        if nombres[i] == buscar:
            equipos[i] = input("Nuevo equipo: ")
            problemas[i] = input("Nuevo problema: ")
            print("Datos modificados")
            return
        i = i + 1

    print("No encontrado")


# Función borrar
def borrarSoporte():
    buscar = input("Nombre a borrar: ")

    i = 0

    while i < len(nombres):
        if nombres[i] == buscar:
            nombres.pop(i)
            equipos.pop(i)
            problemas.pop(i)
            print("Reporte borrado")
            return
        i = i + 1

    print("No encontrado")


# Menú principal
opcion = 0

while opcion != 7:
    print("\n--- SOPORTE TÉCNICO ---")
    print("1. Agregar")
    print("2. Leer")
    print("3. Listar")
    print("4. Buscar")
    print("5. Modificar")
    print("6. Borrar")
    print("7. Salir")

    opcion = int(input("Elige opción: "))

    if opcion == 1:
        agregarSoporte()
    elif opcion == 2:
        leerSoporte()
    elif opcion == 3:
        listarSoporte()
    elif opcion == 4:
        buscarSoporte()
    elif opcion == 5:
        modificarSoporte()
    elif opcion == 6:
        borrarSoporte()
    elif opcion == 7:
        print("Fin del programa")
    else:
        print("Opción incorrecta")

print("Programa generado por Arevalo Rafael 0031")
```
