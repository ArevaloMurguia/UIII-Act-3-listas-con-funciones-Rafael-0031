## Soporte tecnico

```python
# Clase SoporteTecnico

class SoporteTecnico:

    # Constructor
    def __init__(self, cliente, equipo, problema):
        self.cliente = cliente
        self.equipo = equipo
        self.problema = problema

    # Mostrar datos
    def mostrarDatos(self):
        print("Cliente:", self.cliente)
        print("Equipo:", self.equipo)
        print("Problema:", self.problema)


# Lista de objetos
listaSoporte = []


# Agregar soporte
def agregarSoporte():
    cliente = input("Cliente: ")
    equipo = input("Equipo: ")
    problema = input("Problema: ")

    nuevo = SoporteTecnico(cliente, equipo, problema)
    listaSoporte.append(nuevo)

    print("Reporte agregado")


# Leer soporte
def leerSoporte():
    cliente = input("Cliente: ")
    equipo = input("Equipo: ")
    problema = input("Problema: ")

    nuevo = SoporteTecnico(cliente, equipo, problema)
    nuevo.mostrarDatos()


# Listar soportes
def listarSoporte():
    if len(listaSoporte) == 0:
        print("No hay reportes")
    else:
        for x in listaSoporte:
            print("------ REPORTE ------")
            x.mostrarDatos()


# Buscar soporte
def buscarSoporte():
    nombre = input("Cliente a buscar: ")

    for x in listaSoporte:
        if x.cliente == nombre:
            x.mostrarDatos()
            return

    print("No encontrado")


# Modificar soporte
def modificarSoporte():
    nombre = input("Cliente a modificar: ")

    for x in listaSoporte:
        if x.cliente == nombre:
            x.equipo = input("Nuevo equipo: ")
            x.problema = input("Nuevo problema: ")
            print("Reporte modificado")
            return

    print("No encontrado")


# Borrar soporte
def borrarSoporte():
    nombre = input("Cliente a borrar: ")

    for x in listaSoporte:
        if x.cliente == nombre:
            listaSoporte.remove(x)
            print("Reporte borrado")
            return

    print("No encontrado")


# Menu principal
opcion = 0

while opcion != 7:
    print("\n--- MENU SOPORTE TECNICO ---")
    print("1. Agregar")
    print("2. Leer")
    print("3. Listar")
    print("4. Buscar")
    print("5. Modificar")
    print("6. Borrar")
    print("7. Salir")

    opcion = int(input("Elige opcion: "))

    if opcion == 1:
        agregarSoporte()
    elif opcion == 2:
        leerSoporte()
    elif opcion == 3:
        listarSoporte()
    elif opcion == 4:
        buscarSoporte()
    elif opcion == 5:
        modificarSoporte()
    elif opcion == 6:
        borrarSoporte()
    elif opcion == 7:
        print("Fin del programa")
    else:
        print("Opcion incorrecta")
```
