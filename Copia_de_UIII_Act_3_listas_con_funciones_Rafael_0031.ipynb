## Soporte tecnico

```python
# Lista
soportes = []

# Agregar
def agregarsoporte():

    cliente = input("Cliente: ")
    equipo = input("Equipo: ")
    problema = input("Problema: ")

    soportes.append([cliente, equipo, problema])

    print("Agregado")


# Leer
def leersoporte():

    cliente = input("Cliente: ")

    for s in soportes:

        if s[0] == cliente:

            print(s)

            return

    print("No encontrado")


# Listar
def listarsoportes():

    for s in soportes:

        print(s)


# Buscar
def buscarsoporte():

    cliente = input("Buscar: ")

    for s in soportes:

        if s[0] == cliente:

            print("Encontrado", s)

            return

    print("No existe")


# Modificar
def modificarsoporte():

    cliente = input("Modificar: ")

    for s in soportes:

        if s[0] == cliente:

            s[0] = input("Nuevo cliente: ")
            s[1] = input("Nuevo equipo: ")
            s[2] = input("Nuevo problema: ")

            print("Modificado")

            return

    print("No encontrado")


# Borrar
def borrarsoporte():

    cliente = input("Borrar: ")

    for s in soportes:

        if s[0] == cliente:

            soportes.remove(s)

            print("Eliminado")

            return

    print("No encontrado")


# Menú
while True:

    print("\n1.Agregar")
    print("2.Leer")
    print("3.Listar")
    print("4.Buscar")
    print("5.Modificar")
    print("6.Borrar")
    print("7.Salir")

    op = input("Opción: ")

    if op == "1":
        agregarsoporte()

    elif op == "2":
        leersoporte()

    elif op == "3":
        listarsoportes()

    elif op == "4":
        buscarsoporte()

    elif op == "5":
        modificarsoporte()

    elif op == "6":
        borrarsoporte()

    elif op == "7":

        print("Programa realizado por Rafael Arevalo 2-H")

        break

    else:
        print("Error")
```
